```python
import pandas as pd
```

```python
tiatanic=pd.read_csv("F:\Downloads\downoad R\Titanic.csv")
```

```python
tiatanic
```

```python
tia=tiatanic.drop(["name","sibsp","ticket","cabin","body","home.dest","boat"],axis=1)
```

```python
tia.columns
```

```python
tia
```

```python
tia.sex=tia.sex.replace({"male":1,"female":0})
```

```python
tia.embarked=tia.embarked.replace({"S":0,"C":1,"Q":2})
```

```python
ag=tia.age.isnull()
ag.value_counts()
```

```python
tia.age.mean()
```

```python
tia.age=tia.age.fillna('29.842592624521075')
```

```python
tia=tia.dropna()
```

```python
tia.isnull().sum()     ###This checks the null values
```

```python
tia.describe(include="all")
```

```python
x=tia.drop("survived",axis=1)
```

```python
y=tia["survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
pred_value=logmodel.predict(x_test)
```

```python
pred_value
```

```python
df=pd.crosstab(pred_value,y_test)
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tab1=confusion_matrix(pred_value,y_test)
```
